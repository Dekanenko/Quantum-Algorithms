```python
from qiskit.circuit import QuantumCircuit
from qiskit import IBMQ, BasicAer
from qiskit.visualization import plot_histogram
from qiskit import execute

import numpy as np
```

```python
def oracle(n_qubits, balanced=True):
    qc = QuantumCircuit(n_qubits+1)
    
    if balanced:
        for qubit in range(n_qubits):
            qc.cx(qubit, n_qubits)
    else:
        if np.random.randint(2) == 1:
            qc.x(n_qubits)
    
    oracle = qc.to_gate()
    oracle.name = "Oracle"
    
    return oracle
```

```python
def dj_algorithm(n_qubits, balanced=True):
    qc = QuantumCircuit(n_qubits+1, n_qubits)
    
#     prepare ancilla qubit
    qc.x(n_qubits)
    qc.h(n_qubits)
    
#     put qubits in the uniform superposition
    qc.h(range(n_qubits))
    
#     apply oracle
    qc.append(oracle(n_qubits, balanced), range(n_qubits+1))
    
    qc.h(range(n_qubits))
    
    qc.measure(range(n_qubits), range(n_qubits))
    
    return qc
    
```

```python
n_qubits = 4
dj_circuit = dj_algorithm(n_qubits, balanced=True)
# dj_circuit.decompose(reps=2).draw("mpl")
dj_circuit.draw("mpl")
```

```python
n_qubits = 4
dj_circuit = dj_algorithm(n_qubits, balanced=False)

backend = BasicAer.get_backend("qasm_simulator")
shots = 1024
results = execute(dj_circuit, backend=backend, shots=shots).result()
answer = results.get_counts()

plot_histogram(answer)
```
